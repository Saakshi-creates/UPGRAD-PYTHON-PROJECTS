```python
#IMPORTING 
from tkinter import *
from tkinter import messagebox
import random
```

```python
root=Tk()
root.title("LET'S PLAY 'TIC TAC TOE'")

#CREATING VARIABLES TO KEEP TRACK OF THE ONE WHO IS CLICKING
X=True
count=0

#FUNCTION IN ORDER TO DISABLE ALL THE BUTTONS
def disable_all():
    button1.config(state=DISABLED)
    button2.config(state=DISABLED)
    button3.config(state=DISABLED)
    button4.config(state=DISABLED)
    button5.config(state=DISABLED)
    button6.config(state=DISABLED)
    button7.config(state=DISABLED)
    button8.config(state=DISABLED)
    button9.config(state=DISABLED)


#CHECKING IF SOMEONE HAS WON OR NOT
def won_lose():
    global win
    win = False

#HORIZONTAL 1
    if button1["text"]=="X" and button2["text"]=="X" and button3["text"]=="X":
        win=True
        button1.config(bg="blue")
        button2.config(bg="blue")
        button3.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
#HORIZONTAL 2 
    elif button4["text"]=="X" and button5["text"]=="X" and button6["text"]=="X":
        win=True
        button4.config(bg="blue")
        button5.config(bg="blue")
        button6.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
#HORIZONTAL 3
    elif button7["text"]=="X" and button8["text"]=="X" and button9["text"]=="X":
        win=True
        button7.config(bg="blue")
        button8.config(bg="blue")
        button9.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
#VERTICAL 1
    elif button1["text"]=="X" and button4["text"]=="X" and button7["text"]=="X":
        win=True
        button1.config(bg="blue")
        button4.config(bg="blue")
        button7.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
#VERTICAL 2
    elif button2["text"]=="X" and button5["text"]=="X" and button8["text"]=="X":
        win=True
        button2.config(bg="blue")
        button5.config(bg="blue")
        button8.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
#VERTICAL 3
    elif button3["text"]=="X" and button6["text"]=="X" and button9["text"]=="X":
        win=True
        button3.config(bg="blue")
        button6.config(bg="blue")
        button9.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
#DIAGONAL 1
    elif button1["text"]=="X" and button5["text"]=="X" and button9["text"]=="X":
        win=True
        button1.config(bg="blue")
        button5.config(bg="blue")
        button9.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
#DIAGONAL 2       
    elif button3["text"]=="X" and button5["text"]=="X" and button7["text"]=="X":
        win=True
        button3.config(bg="blue")
        button5.config(bg="blue")
        button7.config(bg="blue")
        messagebox.showinfo("TIC TAC TOE RESULT","CONGRATULATIONS!!! X HAS WON!")
        disable_all()
        
#FOR ALL O'S DO SOMETHING

#CHECKING IF THERE'S A TIE
    if count==9 and win==False:
        messagebox.info("TIC TAC TOE", "SEEMS LIKE IT'S A TIE!")
        disable_all()
#CREATING THE CLICKING FUNCTION FOR SELECTING THAT GRID
def clicking(b):
    global X,count
    
    if b["text"] == " " and X == True:
        b["text"] = "X"
        X=False
        count += 1
        won_lose()
    elif b["text"]==" " and X==False:
        b["text"]="O"
        X=True
        count += 1
        won_lose()
    else:
        messagebox.showerror("TIC TAC TOE","THE BOX IS ALREADY SELECTED \n KINDLY SELECT ")
    

#BUILDING THE DIFFERENT GRIDS FOR THE 3*3 SIZED TIC TAC TOE GAME BUTTONS
def reset():
    global button1,button2,button3,button4,button5,button6,button7,button8,button9,X,count
    X=True
    count=0
    button1= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button1)).grid(row=0,column=0)
    button2= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button2)).grid(row=0,column=1)
    button3= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button3)).grid(row=0,column=2)

    button4= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button4)).grid(row=1,column=0)
    button5= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button5)).grid(row=1,column=1)
    button6= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button6)).grid(row=1,column=2)

    button7= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button7)).grid(row=2,column=0)
    button8= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button8)).grid(row=2,column=1)
    button9= Button(root,text=" ", height=4, width=8, bg="SystemButtonFace", command=lambda:clicking(button9)).grid(row=2,column=2)

#RESET GAME
menu1=Menu(root)
root.config(menu=menu1)

options=Menu(menu1,tearoff=False)
menu1.add_cascade(label="Options",menu=options)
options.add_command(label="RESET GAME",command=reset)

reset()
root.mainloop()
```
