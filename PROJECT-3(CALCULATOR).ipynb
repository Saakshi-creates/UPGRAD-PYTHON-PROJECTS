```python
from tkinter import*
import math

def call_result():
    one = int(input_one.get())
    two = int(input_two.get())
    selection = int(operation_num.get())
     
#IF STATEMENTS FOR FINDING THE RESULT AS PER OPERATION
#OPERATION ADDITION
    if selection==1:
        
#ADDING SPACE FOR THE TEXT NOT TO BE SHOWN OVER-WRITTEN BY OTHER OPERATIONS
        Label(root, text = "                                                           ").grid(row=10,column=300)
        result = one+two
        Label(root, text ="Result = %d" %  result).grid(row=10,column=300)
        
#OPEARTION SUBTRACTION        
    elif selection==2:
        Label(root, text = "                                                           ").grid(row=10,column=300)
        result=one-two
        
        Label(root, text ="Result = %d" %  result).grid(row=10,column=300)
        
#OPERATION MULTIPLICATION
    elif selection==3:
        Label(root, text = "                                                                       ").grid(row=10,column=300)
        result=one*two
        
        Label(root, text ="Result = %d" %  result).grid(row=10,column=300)
        
#OPERATION DIVISION
    elif selection==4:
        Label(root, text = "                                                                       ").grid(row=10,column=300)
        if two==0:
            Label(root, text ="DENOMINATOR CANNOT BE ZERO").grid(row=10,column=300)
        else:
            result=one/two
            Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
        
        
#OPERATION HCF
    elif selection==5:
        Label(root, text = "                                                                       ").grid(row=10,column=300)
        smaller=0
        result=0
        if one > two:
            smaller = two
        else:
            smaller = one
        for i in range(1, smaller+1):
            if((one % i == 0) and (two % i == 0)):
                result = i 
                
        Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
       
#OPERATION LCM
    elif selection==6:
        Label(root, text = "                                                                       ").grid(row=10,column=300)
        greater=0
        result=0
        if one > two:
            greater = one
        else:
            greater = two

        while(True):
            if((greater % one == 0) and (greater % two == 0)):
                result = greater
            break
            greater =greater+ 1
        
        Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
        
#OPERATION EXPONENT
    elif selection==7:
        Label(root, text = "                                                                        ").grid(row=10,column=300)
        result= one**two
        
        Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
        
#OPERATION LOGARITHM
    elif selection==8:
        Label(root, text = "                                                                        ").grid(row=10,column=300)
        
        if two==0 or two==1 :
            result=float('inf')
        else:
            result=math.log(one,two)
        Label(root, text ="Result = %d" %  round(result,8)).grid(row=10,column=300)
     
#INVALID OPERAION
    else:
        Label(root, text = "                                                                        ").grid(row=10,column=300)
        Label(root, text = "INVALID OPERATION").grid(row=10,column=300)

        
        
root = Tk()
root.configure(background="#001a33")
input_one =  StringVar()
input_two =  StringVar()
operation_num =  StringVar()

#Definition to view all the available listed commands
def operation():
    op_nm = Label(root, text="INPUT NUMBER 1").grid(row=5, column=150)
    inp_1 = Label(root, text="INPUT NUMBER 2").grid(row=6, column=150)
    inp_2 = Label(root, text="OPERATION NUMBER").grid(row=7, column=150)
    
    
    entryNum1 = Entry(root, textvariable=input_one).grid(row=5, column=270)
    entryNum2 = Entry(root, textvariable=input_two).grid(row=6, column=270)
    entryNum3 = Entry(root, textvariable=operation_num).grid(row=7, column=270) 
    button2=Button(root, text="CLICK TO FIND THE RESULT OF YOUR OPERATION:", command=call_result, fg="#990000", bg="#ccffcc").grid(row=10,column=100)

#Creating label widget

label1=Label(root, text="WELCOME TO GUI BASED CALCULATOR!", fg="#99004d").grid(row=0,column=300)
labelspace=Label(root,text="                  ").grid(row=0,column=10)
label3=Label(root,text="1. ADDITION \n 2. SUBTRACTION \n 3. MULTIPLICATION \n 4. DIVISION \n 5. HCF \n 6. LCM \n 7. POWER \n 8. LOGARITHM").grid(column=100)

labelResult=Label(root)

button1=Button(root, text="ENTER THE NUMBER OF YOUR PREFERRED OPERATION:", command=operation, fg="#0000cc", bg="#ffd9b3").grid(row=3,column=100)


root.mainloop()
```
